### Sequential Workflow


```python
from langgraph.graph import StateGraph , START , END
from langchain_google_genai import ChatGoogleGenerativeAI
from typing import TypedDict 
from dotenv import load_dotenv
```

```python
load_dotenv() 
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
# create a state graph
class LLMState(TypedDict):
    question: str
    answer: str
```

```python
def llm_qa(state: LLMState) -> LLMState:

        # extract a question 
    question = state['question']

        #form a promt
    prompt = f"You are a helpful assistant. If the user greets you, respond politely. Otherwise, answer their asked questions: {question}"

        #ask question    
    answer = model.invoke(prompt).content

        #update the state with the answer
    state['answer'] = answer

    return state 
```

```python
# create a state graph
graph = StateGraph(LLMState)

# add a node
graph.add_node("llm_qa",llm_qa)

# add edges
graph.add_edge(START, "llm_qa")
graph.add_edge("llm_qa", END)

# compile 
workflow = graph.compile()
```

```python
initial_state = LLMState(question="What is the capital of France?", answer="")
fianl_state = workflow.invoke(initial_state)
print(fianl_state)
```
