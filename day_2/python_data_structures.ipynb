### Accessing List Items


**Indexing**

Lists start counting from 0.This is called zero-based indexing.


```python
# this code will print the element 3 which is on index 2 in the list starting the indexes from 0.
numbers = [1, 2, 3, 4, 5]
print(numbers[2])
```

**Negative Indexing** \
In python lists we can count elements from right to left as well , this is called negative indexing .


```python
numbers = [1, 2, 3, 4, 5]
print(numbers[-2])
print(numbers[-1])#if we do not know the extat number of items in a list we can check thet by using -1 index .
```

### **List slicing**

list slicing means extracting a specific portion from the list. Starting index is included while the end is excluded .


```python
numbers = [1, 2, 3, 4, 5]
print(numbers[2:4]) 
# the items on index 2 and 3 are included while index 4 is excluded
```

### **Sep slicing**

Step slicing means how many positions to jump .


```python
numbers = [1, 2, 3, 4, 5]
print(numbers[::2]) 
# it will pick every second item from the list
```

### **List Reversel**


```python
numbers = [1, 2, 3, 4, 5]
print(numbers[::-1]) # this will print all the elemnts from last index to first.
```

#### Task 1: Student Management System


```python
# Creating a list of at least 5 student names
students = ["Ali", "Sara", "Ahmed", "Fatima", "Zainab"]
print()

# Displaying the original list
print("Original Student List:")
print(students)
print()

# Add a new student using append()

students.append("Usman")
print("Studendt List After append().")
print(students)
print()

# Adds a new student using insert()
students.insert(2, "Ayesha")
print("After insert(2, 'Ayesha'):")
print(students)
print()


# Remove a student using remove()
students.remove("Ahmed")
print("After remove('Ahmed'):")
print(students)
print()


#Remove a student using pop()
removed_student = students.pop(3)
print("Student removed using pop():", removed_student)
print("Updated Student List:")
print(students)
print()


#  Creates a sliced list containing only the first 3 students

first_three_students = students[:3]
print("First 3 Students (Sliced List):")
print(first_three_students)
print()


# Sort the list alphabetically
students.sort()
print("Students Sorted Alphabetically:")
print(students)
```

### Task 2: Grade Tracker with Lists


```python
# Grade Tracker with Lists

# Creating two parallel lists
student_names = ["Ali", "Sara", "Ahmed", "Fatima", "Usman"]
student_grades = [85, 72, 91, 48, 65]

# Displaying the original data
print("Student Names:", student_names)
print("Student Grades:", student_grades)

print("----- Grade Report -----")

# Highest grade and student name
highest_grade = max(student_grades)
highest_index = student_grades.index(highest_grade)

print("Highest Grade:")
print(student_names[highest_index], ":", highest_grade)

# Lowest grade and student name
lowest_grade = min(student_grades)
lowest_index = student_grades.index(lowest_grade)

print("\nLowest Grade:")
print(student_names[lowest_index], ":", lowest_grade)


# Students who passed
print("\nStudents Who Passed (Grade >= 50):")

for i in range(len(student_grades)):
    if student_grades[i] >= 50:
        print(student_names[i], ":", student_grades[i])
```
