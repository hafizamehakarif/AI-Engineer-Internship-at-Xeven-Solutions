### Task 1: Student Grade Manager


```python
# Student Grade Manager

# Two parallel lists
names = ["Alice", "Bob", "Carol", "Dave", "Eve"]
grades = [88, 72, 95, 61, 79]


# Add a student
def add_student(name, grade):
    names.append(name)
    grades.append(grade)


# Remove a student
def remove_student(name):
    index = names.index(name)

    names.pop(index)
    grades.pop(index)

# Update a student's grade
def update_grade(name, new_grade):
    index = names.index(name)

    grades[index] = new_grade


# Calculate average grade

def get_average():
    return sum(grades) / len(grades)

# Running the functions

add_student("Frank", 83)
remove_student("Bob")
update_grade("Dave", 74)

average = get_average()

# Highest Grade

highest_grade = max(grades)
highest_index = grades.index(highest_grade)

print("Highest Grade:")
print(names[highest_index], ":", highest_grade)

# Lowest Grade

lowest_grade = min(grades)
lowest_index = grades.index(lowest_grade)

print("\nLowest Grade:")
print(names[lowest_index], ":", lowest_grade)

# Average Grade

print("\nAverage Grade:")
print(average)

# Display All Students

print("\nAll Students:")

for i in range(len(names)):
    print(names[i], ":", grades[i])

# Top 3 Performers

print("\nTop 3 Performers:")

temp_names = names.copy()
temp_grades = grades.copy()

for i in range(3):
    highest = max(temp_grades)

    index = temp_grades.index(highest)

    print(temp_names[index], ":", highest)

    temp_names.pop(index)
    temp_grades.pop(index)

# Students Above Average

print("\nStudents Above or Equal to Average:")

for i in range(len(names)):
    if grades[i] >= average:
        print(names[i], ":", grades[i])

# Students Below Average

print("\nStudents Below Average:")

for i in range(len(names)):
    if grades[i] < average:
        print(names[i], ":", grades[i])
```

### Task 2: Shopping Cart System


```python
# Shopping Cart System

# Parallel lists
item_names = []
item_prices = []
item_quantities = []


# Add an item
def add_item(name, price, quantity):
    item_names.append(name)
    item_prices.append(price)
    item_quantities.append(quantity)


# Remove an item
def remove_item(name):
    index = item_names.index(name)

    item_names.pop(index)
    item_prices.pop(index)
    item_quantities.pop(index)


# Update quantity
def update_quantity(name, new_quantity):
    index = item_names.index(name)

    item_quantities[index] = new_quantity


# Calculate total
def calculate_total():
    total = 0

    for i in range(len(item_names)):
        subtotal = item_prices[i] * item_quantities[i]
        total = total + subtotal

    return total


# --------------------------
# Perform Operations
# --------------------------

add_item("Milk", 20, 2)
add_item("Bread", 15, 3)
add_item("Eggs", 10, 2)
add_item("Butter", 25, 1)
add_item("Juice", 30, 2)

remove_item("Eggs")

update_quantity("Bread", 4)


# --------------------------
# Display Receipt
# --------------------------

print("===== ITEMIZED RECEIPT =====")

for i in range(len(item_names)):
    subtotal = item_prices[i] * item_quantities[i]

    print(
        item_names[i],
        "| Quantity:", item_quantities[i],
        "| Price: $", item_prices[i],
        "| Subtotal: $", subtotal
    )


# --------------------------
# Calculate Final Total
# --------------------------

total = calculate_total()

print("\nTotal Before Discount: $", total)


# Apply Discount
if total > 100:
    discount = total * 0.10
    final_total = total - discount

    print("Discount Applied (10%): $", discount)

else:
    discount = 0
    final_total = total


print("Final Total: $", final_total)


# --------------------------
# Recently Added Items
# --------------------------

print("\nRecently Added Items:")

recent_items = item_names[-3:]

for item in recent_items:
    print(item)
```

### Task 3: Data Cleaning Pipeline


```python
# Data Cleaning Pipeline

# Messy data
data = [
    " Alice ",
    "bob",
    "ALICE",
    None,
    " Carol",
    "bob ",
    "DAVE",
    None,
    "carol"
]

print("Original Data:")
print(data)

# Data Quality Metrics (Before)

total_items = len(data)

non_none_count = len([item for item in data if item is not None])

completeness = (non_none_count / total_items) * 100

# Step 1: Remove None values

without_none = [item for item in data if item is not None]

# Step 2: Remove extra spaces

stripped_data = [item.strip() for item in without_none]

# Step 3: Normalize case 

normalized_data = [item.lower() for item in stripped_data]

# Step 4: Remove duplicates

clean_data = []

for item in normalized_data:
    if item not in clean_data:
        clean_data.append(item)

# Data Quality Metrics (After)

unique_count = len(clean_data)

# Display Results

print("\nCleaned Data:")
print(clean_data)

print("\n----- Data Quality Report -----")

print("Total Items (Before):", total_items)

print("Non-None Items:", non_none_count)

print("Completeness: {:.2f}%".format(completeness))

print("Unique Items After Cleaning:", unique_count)
```
