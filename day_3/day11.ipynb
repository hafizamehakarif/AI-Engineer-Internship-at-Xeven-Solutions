### Task 1


```python
import random

def generate_sample_records(count):
    
    records = []
    for i in range(count):
        if random.random() >0.05:
           records.append({"id":i ,"value":None})
        elif random.random()>0.01:
           records.append({"id":i , "value":"CRITICAL ERROR"})
        else:
             records.append({"id":i , "value":random.randint(0,100)})
    return records

def process_records(records):
   
    processed = []
    total = len(records)

    for index, record in enumerate(records, start=1):
        # Track progress
        if index % 200 == 0:
            print(f"Processing record #{index} of {total}")

        # Stop completely on critical error
        if record["value"] == "CRITICAL_ERROR":
            print(f"CRITICAL ERROR at record {record['id']}. Stopping.")
            break

        # Skip invalid (None) records
        if record["value"] is None:
            continue

        # Apply transformation: double the value
        transformed_value = record["value"] * 2
        processed.append({"id": record["id"], "value": transformed_value})

    return processed


def combine_data_sources(names, scores):
  
    if len(names) != len(scores):
        print("Warning: Data sources have different lengths!")
        print(f"Names: {len(names)}, Scores: {len(scores)}")

    combined = list(zip(names, scores))
    return combined


# --- Run the pipeline ---
records = generate_sample_records(1000)
result = process_records(records)
print(f"\nTotal records processed successfully: {len(result)}")

# Combine data sources example
student_names = ["Ali", "Sara", "Bilal", "Hina"]
student_scores = [85, 92, 78, 88]
combined = combine_data_sources(student_names, student_scores)
print(f"\nCombined data: {combined}")
```

### Day 11 - Task 2: Pattern Generators


```python
def print_multiplication_table(number, limit=10):
   
    print(f"\nMultiplication Table of {number}:")
    for i in range(1, limit + 1):
        print(f"{number} x {i} = {number * i}")


def print_pyramid(rows):

    print(f"\nPyramid with {rows} rows:")
    for i in range(1, rows + 1):
        spaces = " " * (rows - i)
        stars = "*" * (2 * i - 1)
        print(spaces + stars)


def print_number_triangle(rows):
   
    print(f"\nNumber Triangle with {rows} rows:")
    for i in range(1, rows + 1):
        for j in range(1, i + 1):
            print(j, end=" ")
        print()  # newline after each row


def matrix_transpose(matrix):
  
    rows = len(matrix)
    cols = len(matrix[0])
    transposed = []

    for col in range(cols):
        new_row = []
        for row in range(rows):
            new_row.append(matrix[row][col])
        transposed.append(new_row)

    return transposed


def matrix_row_col_sums(matrix):
    
    row_sums = [sum(row) for row in matrix]

    col_sums = []
    for col in range(len(matrix[0])):
        col_total = 0
        for row in range(len(matrix)):
            col_total += matrix[row][col]
        col_sums.append(col_total)

    return row_sums, col_sums


def matrix_diagonal(matrix):
   
    diagonal = []
    for i in range(len(matrix)):
        diagonal.append(matrix[i][i])
    return diagonal


def ascii_art_box(width, height):
  
    print(f"\nASCII Art Box ({width}x{height}):")
    for row in range(height):
        for col in range(width):
            if row == 0 or row == height - 1:
                print("*", end="")  # top/bottom border
            elif col == 0 or col == width - 1:
                print("*", end="")  # side borders
            else:
                print(" ", end="")  # empty inside
        print()  # newline


# --- Run examples ---
print_multiplication_table(7, 5)
print_pyramid(5)
print_number_triangle(5)

sample_matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(f"\nOriginal Matrix: {sample_matrix}")
print(f"Transposed Matrix: {matrix_transpose(sample_matrix)}")

row_sums, col_sums = matrix_row_col_sums(sample_matrix)
print(f"Row sums: {row_sums}")
print(f"Column sums: {col_sums}")
print(f"Diagonal: {matrix_diagonal(sample_matrix)}")

ascii_art_box(10, 5)
```

### Task 3: Number Analysis System


```python

import random

def find_primes_up_to(limit):
    """
    Find all prime numbers up to a given limit.

    Uses an optimized check: only test divisibility up to
    the square root of the number.

    Parameters:
        limit (int): Upper bound (inclusive).

    Returns:
        list: List of prime numbers.
    """
    primes = []
    num = 2

    while num <= limit:
        is_prime = True
        divisor = 2

        # Optimization: only check up to sqrt(num)
        while divisor * divisor <= num:
            if num % divisor == 0:
                is_prime = False
                break
            divisor += 1

        if is_prime:
            primes.append(num)

        num += 1

    return primes


def calculate_factorial(n):
    """
    Calculate factorial of n using an iterative approach.

    Parameters:
        n (int): A non-negative integer.

    Returns:
        int: Factorial of n (n!).
    """
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result


def fibonacci_sequence(count):
    """
    Generate first 'count' numbers of the Fibonacci sequence.

    Parameters:
        count (int): How many numbers to generate.

    Returns:
        list: Fibonacci sequence.
    """
    sequence = []
    a, b = 0, 1

    for _ in range(count):
        sequence.append(a)
        a, b = b, a + b

    return sequence


def number_guessing_game(max_attempts=5, lower=1, upper=100):
    """
    A number guessing game with hints and attempt limit.

    Parameters:
        max_attempts (int): Maximum number of guesses allowed.
        lower (int): Lower bound of number range.
        upper (int): Upper bound of number range.

    Returns:
        None
    """
    secret_number = random.randint(lower, upper)
    attempts = 0

    print(f"\nGuess a number between {lower} and {upper}!")
    print(f"You have {max_attempts} attempts.")

    while attempts < max_attempts:
        try:
            guess = int(input("Your guess: "))
        except ValueError:
            print("Please enter a valid number.")
            continue

        attempts += 1

        if guess == secret_number:
            print(f"Correct! You guessed it in {attempts} attempts!")
            break
        elif guess < secret_number:
            print("Too low! Try a higher number.")
        else:
            print("Too high! Try a lower number.")

        remaining = max_attempts - attempts
        if remaining > 0:
            print(f"Attempts remaining: {remaining}")
    else:
        print(f"Game over! The number was {secret_number}.")


# --- Run examples ---
primes = find_primes_up_to(50)
print(f"Primes up to 50: {primes}")

print(f"Factorial of 5: {calculate_factorial(5)}")
print(f"Factorial of 10: {calculate_factorial(10)}")

fib = fibonacci_sequence(10)
print(f"First 10 Fibonacci numbers: {fib}")

# Uncomment to play interactively:
# number_guessing_game()
```
