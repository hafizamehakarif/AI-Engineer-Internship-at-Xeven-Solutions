## Task 1: Geographic Coordinates System


```python
# Geographic Coordinates System
# This program demonstrates the use of tuples, tuple unpacking,
# functions, and tuple immutability.

import math

# Storing city locations as tuples
# Format: (city_name, latitude, longitude)

cities = [
    ("Islamabad", 33.6844, 73.0479),
    ("Lahore", 31.5204, 74.3587),
    ("Karachi", 24.8607, 67.0011),
    ("Peshawar", 34.0151, 71.5249)
]


# Function to calculate distance between two city tuples
def calculate_distance(city1, city2):
    # Tuple unpacking
    name1, lat1, lon1 = city1
    name2, lat2, lon2 = city2

    # Using a simple distance formula
    distance = math.sqrt((lat2 - lat1) ** 2 + (lon2 - lon1) ** 2)

    return distance


# Function to find the closest city to a given coordinate
def find_closest_city(target_coordinate, cities):
    closest_city = None
    shortest_distance = None

    for city in cities:
        distance = calculate_distance(target_coordinate, city)

        if shortest_distance is None or distance < shortest_distance:
            shortest_distance = distance
            closest_city = city

    # Return city name and distance as a tuple
    return closest_city[0], shortest_distance


# Example target coordinate
target = ("Target Location", 32.0000, 73.0000)

# Find the closest city
city_name, distance = find_closest_city(target, cities)

print(f"The closest city is {city_name}.")
print(f"Distance: {distance:.2f}")


# Demonstrating tuple immutability
print("\nDemonstrating tuple immutability:")

try:
    cities[0][1] = 34.0000
except TypeError as error:
    print("Error:", error)
```

### Task2. Unique Visitor Tracker


```python
# Unique Visitor Tracker

# Website visitors represented by IP addresses
day1_visitors = {
    "192.168.1.1",
    "192.168.1.2",
    "192.168.1.3",
    "192.168.1.4",
    "192.168.1.1"   # Duplicate, automatically removed
}

day2_visitors = {
    "192.168.1.2",
    "192.168.1.3",
    "192.168.1.5",
    "192.168.1.6"
}

print("Day 1 Visitors:", day1_visitors)
print("Day 2 Visitors:", day2_visitors)

# Common visitors
common_visitors = day1_visitors.intersection(day2_visitors)
print("\nCommon Visitors:", common_visitors)

# Visitors only on Day 1
unique_day1 = day1_visitors.difference(day2_visitors)
print("Unique Day 1 Visitors:", unique_day1)

# Visitors only on Day 2
unique_day2 = day2_visitors.difference(day1_visitors)
print("Unique Day 2 Visitors:", unique_day2)

# Total unique visitors
total_visitors = day1_visitors.union(day2_visitors)
print("Total Unique Visitors:", total_visitors)

# Growth Rate
growth_rate = (
    (len(day2_visitors) - len(day1_visitors))
    / len(day1_visitors)
) * 100

print(f"\nVisitor Growth Rate: {growth_rate:.2f}%")

# Retention Rate
retention_rate = (
    len(common_visitors)
    / len(day1_visitors)
) * 100

print(f"Visitor Retention Rate: {retention_rate:.2f}%")
```

### Task3. Email Validation System


```python
# Email Validation System

# Valid email domains
valid_domains = {
    "gmail.com",
    "yahoo.com",
    "outlook.com",
    "hotmail.com"
}

# Registered email addresses
registered_emails = set()

while True:
    email = input("\nEnter email address (or type 'exit' to stop): ")

    if email.lower() == "exit":
        break

    # Check for @ symbol
    if "@" not in email:
        print("Invalid email: Missing @ symbol.")
        continue

    username, domain = email.split("@")

    # Check domain validity
    if domain not in valid_domains:
        print("Invalid email domain.")
        continue

    # Prevent duplicates
    if email in registered_emails:
        print("Email already registered.")
    else:
        registered_emails.add(email)
        print("Email registered successfully.")

print("\nRegistered Emails:")
print(registered_emails)

# Find emails from a specific domain
search_domain = input("\nEnter a domain to search: ")

domain_emails = {
    email
    for email in registered_emails
    if email.endswith("@" + search_domain)
}

print(f"Emails from {search_domain}:")
print(domain_emails)
```
