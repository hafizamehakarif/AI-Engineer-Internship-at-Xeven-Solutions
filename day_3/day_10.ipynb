### Task 1: Student Information System


```python
import json


def add_student(students, student_id, name, age):

    students[student_id] = {
        "name": name,
        "age": age,
        "grades": {}
    }
    return students


def update_grade(students, student_id, subject, score):
    
    if student_id not in students:
        print(f"Student {student_id} not found.")
        return False

    students[student_id]["grades"][subject] = score
    return True


def get_student_average(students, student_id):
    
    student = students.get(student_id)

    if student is None:
        return 0

    grades = student["grades"]

    if not grades:
        return 0

    return sum(grades.values()) / len(grades)


def find_top_student(students):
    
    top_student_id = None
    top_average = -1

    for student_id in students:
        average = get_student_average(students, student_id)

        if average > top_average:
            top_average = average
            top_student_id = student_id

    if top_student_id is None:
        return None

    top_name = students[top_student_id]["name"]
    return (top_student_id, top_name, top_average)


def save_to_json(students, filename="students.json"):
   
    with open(filename, "w") as file:
        json.dump(students, file, indent=4)
    print(f"Data saved to {filename}")


def load_from_json(filename="students.json"):
   
    try:
        with open(filename, "r") as file:
            return json.load(file)
    except FileNotFoundError:
        print(f"{filename} not found. Starting fresh.")
        return {}


def generate_report(students):
   
    print("\n" + "=" * 40)
    print("STUDENT PERFORMANCE REPORT")
    print("=" * 40)

    # Build list of (name, gpa) tuples for sorting
    report_data = []
    for student_id, info in students.items():
        gpa = get_student_average(students, student_id)
        report_data.append((info["name"], student_id, gpa))

    # Sort by GPA descending
    report_data.sort(key=lambda entry: entry[2], reverse=True)

    for rank, (name, student_id, gpa) in enumerate(
            report_data, start=1):
        print(f"{rank}. {name} ({student_id}) - GPA: {gpa:.2f}")


# --- Run the system ---
students = {}

students = add_student(students, "S001", "Ali", 20)
students = add_student(students, "S002", "Sara", 21)
students = add_student(students, "S003", "Bilal", 22)

update_grade(students, "S001", "Math", 85)
update_grade(students, "S001", "Physics", 78)
update_grade(students, "S002", "Math", 92)
update_grade(students, "S002", "Physics", 88)
update_grade(students, "S003", "Math", 70)
update_grade(students, "S003", "Physics", 65)

generate_report(students)

top = find_top_student(students)
print(f"\nTop Student: {top[1]} ({top[0]}) with GPA {top[2]:.2f}")

save_to_json(students, "students.json")

# Demonstrate loading
loaded_students = load_from_json("students.json")
print(f"\nLoaded data for {len(loaded_students)} students")
```

### Task 2: Product Inventory Manager


```python
import json


def add_product(inventory, product_id, name, price,
                 quantity, category):
  
    inventory[product_id] = {
        "name": name,
        "price": price,
        "quantity": quantity,
        "category": category
    }
    return inventory


def update_stock(inventory, product_id, quantity_change):
   
    if product_id not in inventory:
        print(f"Product {product_id} not found.")
        return False

    inventory[product_id]["quantity"] += quantity_change

    # Prevent negative stock
    if inventory[product_id]["quantity"] < 0:
        inventory[product_id]["quantity"] = 0

    return True


def search_by_category(inventory, category):
   
    return {
        product_id: info
        for product_id, info in inventory.items()
        if info["category"] == category
    }


def low_stock_alert(inventory, threshold=10):
   
    alerts = []

    for product_id, info in inventory.items():
        if info["quantity"] < threshold:
            alerts.append((product_id, info["name"],
                            info["quantity"]))

    return alerts


def calculate_total_inventory_value(inventory):
   
    total = 0
    for info in inventory.values():
        total += info["price"] * info["quantity"]
    return total


def average_price_per_category(inventory):
   
    category_totals = {}
    category_counts = {}

    for info in inventory.values():
        category = info["category"]
        price = info["price"]

        category_totals[category] = (
            category_totals.get(category, 0) + price
        )
        category_counts[category] = (
            category_counts.get(category, 0) + 1
        )

    averages = {
        category: category_totals[category] / category_counts[category]
        for category in category_totals
    }

    return averages


def export_inventory_report(inventory, filename="inventory_report.json"):
    """
    Export a full inventory report as JSON, including
    calculated statistics.

    Parameters:
        inventory (dict): The inventory dictionary.
        filename (str): Output filename.

    Returns:
        None
    """
    report = {
        "products": inventory,
        "total_value": calculate_total_inventory_value(inventory),
        "average_price_per_category": average_price_per_category(
            inventory),
        "low_stock_items": low_stock_alert(inventory)
    }

    with open(filename, "w") as file:
        json.dump(report, file, indent=4)

    print(f"Inventory report exported to {filename}")


# --- Run the system ---
inventory = {}

inventory = add_product(inventory, "P001", "Laptop", 1200.00, 5,
                         "Electronics")
inventory = add_product(inventory, "P002", "Mouse", 25.00, 50,
                         "Electronics")
inventory = add_product(inventory, "P003", "Desk Chair", 150.00, 8,
                         "Furniture")
inventory = add_product(inventory, "P004", "Notebook", 2.50, 200,
                         "Stationery")

update_stock(inventory, "P001", -2)  # sold 2 laptops
update_stock(inventory, "P002", 20)  # restocked mice

print(f"Total inventory value: ${calculate_total_inventory_value(inventory):.2f}")

print(f"\nElectronics products:")
electronics = search_by_category(inventory, "Electronics")
for pid, info in electronics.items():
    print(f"  {pid}: {info['name']} - ${info['price']}")

print(f"\nAverage price per category:")
averages = average_price_per_category(inventory)
for category, avg_price in averages.items():
    print(f"  {category}: ${avg_price:.2f}")

print(f"\nLow stock alerts (threshold=10):")
alerts = low_stock_alert(inventory, threshold=10)
for product_id, name, qty in alerts:
    print(f"  {name} ({product_id}): only {qty} left!")

export_inventory_report(inventory)
```

### Task 3: Configuration Manager


```python
import json
import os


# Default configuration - used if config.json doesn't exist
# or is missing keys
DEFAULT_CONFIG = {
    "database": {
        "host": "localhost",
        "port": 5432,
        "name": "myapp_db"
    },
    "api_keys": {
        "openai": "",
        "huggingface": ""
    },
    "features": {
        "enable_logging": True,
        "max_retries": 3,
        "debug_mode": False
    }
}


def create_default_config(filename="config.json"):
    with open(filename, "w") as file:
        json.dump(DEFAULT_CONFIG, file, indent=4)
    print(f"Default config created at {filename}")


def load_config(filename="config.json"):

    if not os.path.exists(filename):
        print(f"{filename} not found. Creating default config.")
        create_default_config(filename)

    try:
        with open(filename, "r") as file:
            return json.load(file)
    except json.JSONDecodeError:
        print(f"{filename} is corrupted. Using defaults.")
        return DEFAULT_CONFIG.copy()

def validate_config(config):

    for key, default_value in DEFAULT_CONFIG.items():
        if key not in config:
            print(f"Missing key '{key}'. Using default.")
            config[key] = default_value

    return config


def get_config_value(config, *keys, default=None):

    current = config

    for key in keys:
        if isinstance(current, dict):
            current = current.get(key)
        else:
            return default

        if current is None:
            return default

    return current


def update_config(config, filename="config.json", **updates):

    config.update(updates)

    with open(filename, "w") as file:
        json.dump(config, file, indent=4)

    print(f"Config updated and saved to {filename}")
    return config


# --- Run the system ---
config = load_config("config.json")
config = validate_config(config)

# Safe access using get_config_value
db_host = get_config_value(config, "database", "host")
db_port = get_config_value(config, "database", "port")
print(f"Database: {db_host}:{db_port}")

# Accessing a key that might not exist - with default
cache_enabled = get_config_value(
    config, "features", "enable_cache", default=False)
print(f"Cache enabled: {cache_enabled}")

# Update an API key
config = update_config(
    config,
    "config.json",
    api_keys={
        "openai": "your-key-here", 
        "huggingface": "your-hf-key"
    }
)
```
