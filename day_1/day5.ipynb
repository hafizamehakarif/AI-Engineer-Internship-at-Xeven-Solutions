```python
# Decision Tree Simulation for Loan Approval

# Get applicant information
age = int(input("Enter your age: "))
income = float(input("Enter your annual income: "))
credit_score = int(input("Enter your credit score: "))

# Root node: Check applicant's age
print("\nDecision Path:")

if age < 18:
    print("Age is less than 18.")
    print("Loan Rejected.")

# Second node: Check income
elif income < 30000:
    print("Age requirement passed.")
    print("Income is less than 30000.")
    print("Loan Rejected.")

# Third node: Check credit score
elif credit_score < 600:
    print("Age requirement passed.")
    print("Income requirement passed.")
    print("Credit score is less than 600.")
    print("Loan Rejected.")

# Leaf node: Approve loan
else:
    print("Age requirement passed.")
    print("Income requirement passed.")
    print("Credit score requirement passed.")
    print("Loan Approved.")
```
