#### Conditional statement


```python
marks = 85

if marks >= 90:
    print("Grade A")
elif marks >= 80:
    print("Grade B")
elif marks >= 70:
    print("Grade C")
else:
    print("Grade D")
```

## Multi-Operation Calculator


```python
# Multi-Operation Calculator

first_number = float(input("Enter the first number: "))
second_number = float(input("Enter the second number: "))

operation = input("Choose an operation (+, -, *, /, %, **): ")

if operation == "+":
    answer = first_number + second_number
    print(first_number, "+", second_number, "=", answer)

elif operation == "-":
    answer = first_number - second_number
    print(first_number, "-", second_number, "=", answer)

elif operation == "*":
    answer = first_number * second_number
    print(first_number, "*", second_number, "=", answer)

elif operation == "/":
    if second_number == 0:
        print("Cannot divide by zero.")
    else:
        answer = first_number / second_number
        print(first_number, "/", second_number, "=", answer)

elif operation == "%":
    if second_number == 0:
        print("Cannot perform modulus with zero.")
    else:
        answer = first_number % second_number
        print(first_number, "%", second_number, "=", answer)

elif operation == "**":
    answer = first_number ** second_number
    print(first_number, "**", second_number, "=", answer)

else:
    print("Invalid operation. Please choose from +, -, *, /, %, or **.")
```
